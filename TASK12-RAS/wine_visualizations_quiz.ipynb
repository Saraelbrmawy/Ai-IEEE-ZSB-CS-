# Plotting with Matplotlib
Use Matplotlib to create bar charts that visualize the conclusions you made with groupby and query.

```python
# Import necessary packages and load `winequality_edited.csv`
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('winequality_edited1.csv')
df
```

### #1: Do wines with higher alcoholic content receive better ratings?
Create a bar chart with one bar for low alcohol and one bar for high alcohol wine samples.

```python
# Use query to select each group and get its mean 
low_alcohol= df.query('alcohol <= 10.5')['quality'].mean()
high_alcohol = df.query('alcohol > 10.5')['quality'].mean()

```

```python
# Create a bar chart with proper labels
low_alcohol= df.query('alcohol <= 10.5')['quality'].mean()
high_alcohol = df.query('alcohol > 10.5')['quality'].mean()
fig, ax = plt.subplots()
ax.bar(['low', 'high'], [low_alcohol, high_alcohol ])
ax.set_xlabel('Alcohol Content')
ax.set_ylabel('Mean Quality')
ax.set_title('Mean Quality of Low and High Alcohol Wines')
plt.show()
```

### #2: What level of acidity receives the highest average rating?
Create a bar chart with a bar for each of the four acidity levels.

```python
# Use groupby to get the mean quality for each acidity level
mean_df=df.groupby('acidity_level')['quality'].mean()
mean_df
```

```python
# Create a bar chart with proper labels
fig,ax=plt.subplots()
ax.bar(mean_df.index,mean_df.values)
ax.set_xlabel('Acidity Level')
ax.set_ylabel('Mean Quality')
ax.set_title('Meann Quality of wines by acidity level')
plt.show()
```
